# HW1 generated images viewer

This notebook:
1. Reads `hw1_script.py`
2. Extracts the PNG filenames from lines **444–453** (1-indexed)
3. Displays each image, preceded by its filename

Run the cells below from the **same directory** as `hw1_script.py` and the generated PNGs.


```python
from pathlib import Path
import re

SCRIPT = Path('hw1_script.py')
assert SCRIPT.exists(), f"Couldn't find {SCRIPT.resolve()}"

# Extract PNG names from lines 444–453 (1-indexed)
start_line, end_line = 444, 453
lines = SCRIPT.read_text(encoding='utf-8').splitlines()
snippet = "\n".join(lines[start_line-1:end_line])

pngs = re.findall(r"['\"]([^'\"]+\.png)['\"]", snippet)
pngs = list(dict.fromkeys(pngs))  # de-dup, preserve order

print('Found PNGs:')
for p in pngs:
    print(' -', p)

assert len(pngs) > 0, (
    "No .png filenames found in lines 444–453. "
    "Double-check the line numbers and that the script matches your local copy."
)
```

```python
from IPython.display import display, Markdown, Image
from pathlib import Path

for fname in pngs:
    display(Markdown(f"## `{fname}`"))
    path = Path(fname)
    if path.exists():
        display(Image(filename=str(path)))
    else:
        display(Markdown(f"⚠️ **File not found:** `{fname}` (expected at {path.resolve()})"))
```
